```python

corpus = corpus_loader.load(corpus_filepath="./corpus-v0.4b/tbmm_corpus.mm")
```

```python
lda, topic_dist_matrix, label_vector = corpus_loader.load_lda_model(corpus, "./corpus-v0.4b/tbmm_corpus.mm.tbmm_lda.model")
```

```python
lda.print_topics()
```

To obtain a plot of the topic distributions of 4, 9, 12, 15 and 16 as in the paper, run the following cell.

```python
# not available for version 4.0, for now.

for start in [0, 5, 10, 15]:
    corpus.plot_a_specific_topic_by_year(range(start, start+5), 
                                                       topic_dist_matrix, 
                                                       label_vector, 
                                                       ["Topic %d" % (i+1) for i in range(start, start+5)], 
                                                       keyword="DEMO_topics_start")
```
